```python
from mpl_toolkits import mplot3d  #provides some basic 3D plotting (scatter, surf, line, mesh) tools.
```

<p>Here the class and its respective methods are defined, so that it can be instantiated later and it can build some 3d objects according to certain parameters .</p>

```python

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math as mt
x = np.linspace(-6, 6, 60)
y = np.linspace(-6, 6, 60)

class Solid3d(object):
# Constructor
    def __init__(self, x, y, s):
                self.x = x
                self.y = y
                self.s = s
# Methods                
    def sigmoide(self,x):  
        return np.exp(-np.logaddexp(0, -x)) 
#This method create some kinds of solids 3d according with its mathematical functions..

    def f(self,x,y,s):
        r=0
        if (s=='Ripple'):    
            r=np.sin(np.sqrt(x ** 2 + y ** 2))
        if (s=='Pyramid'):
            r=(1-np.abs(x+y)-np.abs(y-x)) 
        if (s=='Sheet'):
            r=(((self.sigmoide(-.65-x)+self.sigmoide(-.35-x)+self.sigmoide(-.05-x)+self.sigmoide(.25-x)+self.sigmoide(.55-x))/7))
        if (s=='Windwhille'):
            r=(self.sigmoide(x*y)*self.sigmoide(1-(x*9)**2+(y*9)**2)/9)
        if (s=='Mountain'):
            r=((self.sigmoide(0.2-(x**2+y**2)) + self.sigmoide(0.2-(x**2/3+y**2/3)))/3-1) 
        if (s=='Roof'):
            r=(1-np.abs(y))
        return(r)   
   #This method allows visualization of each 3D solid.. 
   #title:title of the solid
   #xy:Angle in xy plane
   #xz:Angle in xz plane
   #rstride:distance longitudinal stride 
   #cstride:distance latitudinal stride  
    def render(self,title,xy,xz,rstride,cstride,cmap,edgecolor):    
            ax = plt.axes(projection='3d')
            ax.set_title(title)        
            ax.view_init(xy, xz)
            X, Y = np.meshgrid(self.x, self.y)
           # Z = self.f(X, Y, s)
            Z = self.f(X, Y, title)
            ax.plot_surface(X, Y, Z, rstride=rstride, cstride=cstride,
             cmap=cmap, edgecolor=edgecolor);
            


```

<b>Ripple: $f(x,y)=sin(\sqrt{x^2 + y^2})$<b>


```python
title = 'Ripple'
sd3d=Solid3d(x,y,title)
sd3d.f(x,y,title)
xy = 45
xz = 25
rstride = 2
cstride = 2
cmap='viridis'
edgecolor='blue'
sd3d.render(title,xy,xz,rstride,cstride,cmap,edgecolor)
```

<b>Pyramid: $f(x,y)=(1-abs(x+y)-abs(y-x))$<b>

```python
title = 'Pyramid'
sd3d=Solid3d(x,y,title)
sd3d.f(x,y,title)
xy = 35
xz = 25
rstride = 2
cstride = 2
cmap='viridis'
edgecolor='blue'
sd3d.render(title,xy,xz,rstride,cstride,cmap,edgecolor)

```
